## Part 1:<br>
Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable.<br>
For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable.

```python
# using the read_csv method to read he csv as a dataframe
df=pd.read_csv('nba.csv')
```

```python
df
```

```python
# using the pandas groupby method to group data based on categorical variables 
df.groupby('College')['Age'].describe()
```

```python
# using the pandas groupby method to group data based on categorical variables 
df.groupby(['Team','Position'])['Salary'].describe()
```

```python
bins=[19,26,33,40] # bin array for setting criteria to bin(slice) by while using the cut method:
labels=['19-26','26-33','33-48']

df['Age Category']=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
df.groupby('Age Category')['Salary'].describe()
```
